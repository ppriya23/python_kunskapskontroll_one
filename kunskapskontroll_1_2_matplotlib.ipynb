### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Answer: Plotting in Matplotlib can be done in either of two ways:
        1. Object-Oriented approach
        2. pyplot / MATLAB style
        It is recommended to use Object-Oriented approach, especially for more complicated plots and plots that are part of a bigger project.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Answer: Figure is the whole figure and can be seen in canvas.
        Axes is the part of the "canvas" that the plot is attached on. We call methods that do the plotting directly from the Axes which gives high flexibility in customizing our plots.
        Axis is sets the scale and limits and generate ticks (mark on the axis) and ticklables (string lebelling the ticks).
        Artist is basically everything visible on the figure is an artist, including figure, axes and axis. Most Artist are tied to an axes.

### When plotting in Matplotlib, what is the expected input data type?

Answer: When plotting with Matplotlib, you should general use numpy arrays as input or objects that can be transformed to a numpy array with the function numpy.asarray.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Generate X values using np.linspace
x = np.linspace(-4, 4, 100)  # 100 points from -4 to 4
y = x**2

# when doing the actual plot we call the methods directly from the axes in object_oriented approach
fig, ax = plt.subplots()
ax.plot(x, y, label = 'y = x^2')
ax.set_title('Object_Oriented approach y = x^2') # Add a title to the axes
ax.set_xlabel('x-label') # Add an x_label to the axes
ax.set_ylabel('y-label') # Add an y_label to the axes
ax.legend()
```

```python
# pyplot approach with title and axis lebels
plt.plot(x, y, label = 'y = x^2')
plt.title('pyplot approach y = x^2')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.legend()
```

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())


#creating subplot
fig, axs = plt.subplots(1, 2, figsize = (12, 5))

axs[0].scatter(random_data_x, random_data_y)
axs[0].set_title('Scatter Plot')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')

axs[-1].bar(names,values)
axs[-1].set_title('Bar Plot')
axs[-1].set_xlabel('fruits')
axs[-1].set_ylabel('values')

plt.show()
```
